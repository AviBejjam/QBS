```python
import json
# Path to the JSON file
json_file_path = folder_path + "/output/structuredData.json"

# Path to the text file
text_file_path = folder_path + "/text_files/output_text.txt"
# Word to search for
search_word = '"Text":'

# Open the text file in write mode
with open(text_file_path, 'w') as text_file:
    # Open and read the JSON file line by line
    with open(json_file_path, 'r') as json_file:
        for line in json_file:
            # Check if the line contains the search word
            if search_word in line:
                # Find the index of the search word in the line
                index = line.find(search_word)
                # Write the remaining part of the line (after the search word) to the text file
                text_file.write(line[index + len(search_word):])
```
