```python
import json
import csv
import os
import re
```

```python
current_directory = os.getcwd()
folder_path = os.path.dirname(current_directory)
```

```python

# Directory containing JSON files
json_directory = folder_path + "/output_json"

# Parent directory containing the text_files directory
parent_directory = os.path.dirname(json_directory)

# Directory containing text files
text_directory = os.path.join(parent_directory, "text_files")

# Create the text directory if it doesn't exist
os.makedirs(text_directory, exist_ok=True)

# Regular expression pattern to extract text within double quotes until the comma
pattern = re.compile(r'"Text":\s*"([^"]+)",')

# Loop through each JSON file in the directory
for json_file_name in os.listdir(json_directory):
    if json_file_name.endswith('.json'):
        # Path to the current JSON file
        json_file_path = os.path.join(json_directory, json_file_name)
        
        # Path to the corresponding text file
        text_file_path = os.path.join(text_directory, f"{json_file_name.split('.')[0]}_output_text.txt")

        # Open the text file in write mode
        with open(text_file_path, 'w') as text_file:
            # Open and read the JSON file
            with open(json_file_path, 'r') as json_file:
                # Read the entire JSON file
                data = json_file.read()
                # Find all matches of the pattern in the JSON data
                matches = pattern.findall(data)
                for match in matches:
                    # Write the match to the text file
                    text_file.write(match + '\n')
```

```python
print("Extraction Done")
```
